```python
import numpy as np
# https://www.youtube.com/watch?v=cJ21moQpofY
# video explica bem

num_items = 5 # numero de items totais
capacidade_mochila = 7 # peso maximo que a mochila leva
valor_items = [2,2,4,5,3] # valor de cada item
pesos_items = [3,1,3,4,2] # peso de cada item

'''capacidade_mochila = 20000
valor_items = [random.randint(10,150) for x in range(0,500)]
pesos_items = [random.randint(1,50) for x in range(0,500)]
num_items = 500'''

def mochila(n,c,vl,ps):
    m = np.zeros((n,c))

    for i in range(n):
        for k in range(c):
            if i != 0 and k != 0:
                if ps[i-1] <= k:
                  m[i][k] = max(vl[i-1] + m[i-1][k-ps[i-1]],m[i-1][k])
                else:
                  m[i][k] = m[i-1][k]
    
    return m

m = mochila(num_items+1,capacidade_mochila+1,valor_items,pesos_items)

print(m)
print(np.max(m))
```

```python
capacidade_mochila = 900
valor_items = [random.randint(10,150) for x in range(0,50)]
pesos_items = [random.randint(1,50) for x in range(0,50)]
num_items = 50

# SGD parameters
learning_rate = 0.01
num_iterations = 1000
penalty_param = 10


inicio = time.time()
print("melhor beneficio possivel: "+str(np.max(mochila(num_items+1,capacidade_mochila+1,valor_items,pesos_items))))
fim = time.time()
print("tempo execucao algoritmo normal: "+str(fim-inicio))

def calc_value(x, values, weights, capacity, penalty):
    """
    Objective function with penalty for exceeding weight capacity.
    """
    total_value = np.dot(values, x)
    total_weight = np.dot(weights, x)
    penalty_term = penalty * max(0, total_weight - capacity)
    return total_value - penalty_term

def stochastic_gradient_descent(values, weights, capacity, lr, iterations, penalty):
    # Initialize x (continuous variables between 0 and 1)
    x = np.random.uniform(0, 1, len(values))
    values = np.array(values)
    weights = np.array(weights)
    history = []

    for _ in range(iterations):
        # Randomly select one item to compute the stochastic gradient
        i = random.randint(0, len(values) - 1)

        # Compute the weight violation
        total_weight = np.dot(weights, x)
        weight_violation = max(0, total_weight - capacity)

        # Gradient computation for a single item
        gradient = np.zeros(len(values))
        gradient[i] = values[i] - 2 * penalty * weight_violation * weights[i]

        # Update the solution
        x = x + lr * gradient

        # Ensure x stays in the range [0, 1]
        x = np.clip(x, 0, 1)

        # Save the current objective value for analysis
        obj_value = calc_value(x, values, weights, capacity, penalty)
        history.append(obj_value)

    # Convert x to binary solution
    x_binary = (x >= 0.5).astype(int)

    # Final total value and weight
    total_value = np.dot(values, x_binary)
    total_weight = np.dot(weights, x_binary)

    return total_value, total_weight, x_binary, history

inicio = time.time()
total_value, total_weight, solution, history = stochastic_gradient_descent(
    valor_items, pesos_items, capacidade_mochila, learning_rate, num_iterations, penalty_param
)
fim = time.time()

print(total_value)
print(total_weight)
print("tempo "+str(fim - inicio))
```
